```python
import secrets

import x25519
```

```python
a = int.from_bytes(secrets.token_bytes(32), "little")
A = x25519.mult(a, 9)
print(A)
```

```python
b = int.from_bytes(secrets.token_bytes(32), "little")
B = x25519.mult(b, 9)
print(B)
```

```python
ka = x25519.mult(a, B)
kb = x25519.mult(b, A)

print(ka)
print(kb)
print(ka == kb)
```
